```python
#importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
myData=pd.read_csv("Social_Network_Ads.csv")
myData
```

```python
myDummyData=pd.get_dummies(myData.Gender)
myDummyData
```

```python
joinData=pd.concat([myData,myDummyData],axis='columns')
```

```python
finalData=joinData.drop(['Gender','Female'],axis='columns')
finalData
```

```python
x=finalData.drop(['Purchased'],axis='columns')
x
```

```python
y=finalData.Purchased
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
createModel=LogisticRegression(random_state=0)
myModel=createModel.fit(x_train,y_train)
```

```python
y_prediction=createModel.predict(x_test)
y_prediction
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_prediction)
cm
```
